This is just an initial exploration of the PlantVillage dataset after being split into train and test set.

# 1. Libraries

```python
#Mount Google Drive
from google.colab import drive

drive.mount('/content/drive')  


# Import libraries
import os
import matplotlib.pyplot as plt
import numpy as np
```

# 2. Loading

```python
# Paths
train_dir = "/content/drive/MyDrive/plant_village_dataset/train"
test_dir  = "/content/drive/MyDrive/plant_village_dataset/test"

#  Get class names (subfolder names in train folder)
class_names = sorted(os.listdir(train_dir))

# Count images per class using listdir (fast)
train_counts = {cls: len(os.listdir(os.path.join(train_dir, cls))) for cls in class_names}
test_counts  = {cls: len(os.listdir(os.path.join(test_dir, cls))) for cls in class_names}
```

# 3. Some numbers

```python
print("Number of classes:", len(class_names))
print("Number of images in train set:", sum(train_counts.values()))
print("Number of images in test set:", sum(test_counts.values()))
```

# 4. A graph of the numbers of samples per class

```python
fig, ax = plt.subplots(figsize=(12, 6))
x = np.arange(len(class_names))
width = 0.35

ax.bar(x - width/2, list(train_counts.values()), width, label='Train', color='green', alpha=0.8)
ax.bar(x + width/2, list(test_counts.values()), width, label='Test', color='orange', alpha=0.8)

ax.set_xlabel('Classes', fontsize=12, fontweight='bold')
ax.set_ylabel('Number of Images', fontsize=12, fontweight='bold')
ax.set_title('Train vs Test Data Distribution of PlantVillage Dataset', fontsize=14, fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels(class_names, rotation=90, fontsize=7)
ax.legend()
plt.tight_layout()
plt.show()
```
