```python
import os
import shutil
```

```python
#for train set
plantvillage_train = "/content/drive/MyDrive/plant_village_dataset/train"
plantdoc_train     = "/content/drive/MyDrive/plant_village_dataset/plantdoc_split/train"
```

```python
#for val set
plantvillage_train = "/content/drive/MyDrive/plant_village_dataset/val_downsampled"
plantdoc_train     = "/content/drive/MyDrive/plant_village_dataset/plantdoc_split/val"
```

```python
#  Get class folders 
village_classes = set(os.listdir(plantvillage_train))
doc_classes     = os.listdir(plantdoc_train)

print("PlantVillage classes:", len(village_classes))
print("PlantDoc classes:", len(doc_classes))
```

```python
copied = 0

for cls in doc_classes:
    src_cls_folder = os.path.join(plantdoc_train, cls)

    # skip non-folders (just in case)
    if not os.path.isdir(src_cls_folder):
        continue



    dest_cls_folder = os.path.join(plantvillage_train, cls)

    # ensure folder exists
    os.makedirs(dest_cls_folder, exist_ok=True)

    # copy each image
    for img_name in os.listdir(src_cls_folder):
        src_img = os.path.join(src_cls_folder, img_name)

        # generate new unique filename to avoid overwriting
        base, ext = os.path.splitext(img_name)
        new_name = f"doc_{base}{ext}"   # prefix all PlantDoc images
        dest_img = os.path.join(dest_cls_folder, new_name)

        shutil.copy2(src_img, dest_img)
        copied += 1

print("\n MERGE COMPLETE")
print(f"Copied PlantDoc images: {copied}")
```
