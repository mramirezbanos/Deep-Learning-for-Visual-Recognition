```python
import os
import random
import shutil
```

```python
split_ratio = 0.10  # 10%

pv_train_dir = "/content/drive/MyDrive/plant_village_dataset/train"
pv_val_dir   = "/content/drive/MyDrive/plant_village_dataset/val"

pd_train_dir = "/content/drive/MyDrive/plant_village_dataset/plantdoc_split/train"
pd_val_dir   = "/content/drive/MyDrive/plant_village_dataset/plantdoc_split/val"
```

```python
# Function: make a per-class validation split

def create_val_split(train_dir, val_dir, ratio=0.1):
    os.makedirs(val_dir, exist_ok=True)

    class_names = sorted(os.listdir(train_dir))

    for cls in class_names:
        cls_train_path = os.path.join(train_dir, cls)

        if not os.path.isdir(cls_train_path):
            continue

        cls_val_path = os.path.join(val_dir, cls)
        os.makedirs(cls_val_path, exist_ok=True)

        images = [f for f in os.listdir(cls_train_path)
                  if os.path.isfile(os.path.join(cls_train_path, f))]

        num_val = max(1, int(len(images) * ratio))
        val_images = random.sample(images, num_val)

        print(f"Class: {cls} , Moving {num_val} images to validation set")

        for img in val_images:
            src = os.path.join(cls_train_path, img)
            dst = os.path.join(cls_val_path, img)
            shutil.move(src, dst)

    print(f"\nDone! Validation split created at: {val_dir}\n")
```

```python
# Run for both datasets


print("Splitting PlantVillage")
create_val_split(pv_train_dir, pv_val_dir, split_ratio)

print("Splitting PlantDoc")
create_val_split(pd_train_dir, pd_val_dir, split_ratio)
```
